3번 문제입니다.

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

•1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

range(stop)

range(start, stop[, step])


•2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 

•3) frange 사용 예◾3-1) frange(0.5)◦[0.0, 0.1, 0.2, 0.3, 0.4]

◾3-2) frange(1.0, 2.0)◦[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

◾3-3) frange(2.2, 4.0, 0.5)◦[2.2, 2.7, 3.2, 3.7]




```python
def frange(stop, start = 0, step = 0.1):
    a = 0
    lists = []
    if stop <= 0 or start < 0 or step <= 0:
        print '음수는 받지 않습니다.', 
        return 
    else:
        while start < stop:
            if stop == 0.0:
                lists.append(start)
                start += step
            else:
                lists.append(start)
                start += step
        start = start-step
    return lists
        

print frange(4, 3, 0.5) #stop, start, step순으로 매개변수를 넣어줍니다.
print frange(-4)
```

4번 문제입니다.

가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

•sum() 사용 예 

◾sum()◦0

◾sum(1, 2)◦3

◾sum(1, 2, 3, 4, 5)◦15

◾sum(1, 5, 7, 2, -10)◦5




```python
def sum(s=0,*array):
    for i in array:
        s += i
    return s
print sum(1, 3, 4, 5, 6)
```

5번 문제입니다.

여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 

다음 조건을 참고하여 작성하시오.

•1) 다음에 제시되는 함수들을 모두 이용해야 한다. 

◾split

◾map

◾join

•2) myinitial() 함수 사용 예

◾myinitial("as soon as possible")◦ASAP




```python
def returnfirst(stance):
    return stance[0].upper()
    
def myinitial(distance):
    words = distance.split()
    lists = map(returnfirst, words)
    lists = ''.join(lists)
    return lists

print myinitial("my life for aiur")
```

5번 문제 해설

우선 returnfirst와 myinitial()의 두 개의 함수를 선언한다. 그리고 myinitial()함수를 통해서 입력받은 문자열을 가져온다.

함수 내에서 가져온 문자열을 split()로 분리하여 words 리스트에 저장하고 그 리스트를 앞서 만든 returnfirst()함수를 통해서 맨 앞자리들을

대문자로 바꾸어 가져온다. 그리고 리스트 내의 모든 문자들을 join()을 통하여 하나로 합친다. 

마지막으로 하나로 합쳐진 문자열을 반환한다.

6번 문제입니다.

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

•[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 


```python
def myfact(n):
    if n > 1:
        return n*myfact(n-1)
    else :
        return 1

print myfact(500)
```

6번 문제 해설

myfact() 함수 내에서 인자로 받는 값이 1보다 크면 n*myfact(n-1)을 반환하고

그렇지 않으면 1을 반환하여 재귀호출이 영원히 반복되지 않도록 한다.

8번 문제입니다.

10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.

이백만(2,000,000) 이하 소수의 합은 얼마입니까?

-------------------------------------------------------------------------------------------------------------------

```python
import math

def isPrime(x): 
    returning = True 
    for i in range(2,int(math.sqrt(x))+1):
        if x%i==0: 
            returning = False
            return returning
    return returning

sum = 0
for x in range(2, 2000001):
    if isPrime(x):
        sum += x
print sum
```

우선 isPrime()함수를 통하여 매개변수로 들어간 숫자가 소수인지 아닌지 판별합니다. 

그리고 for문 반복을 이용하여 2부터 2백만까지 모든 소수의 합을 구하여 그 합을 출력합니다.

10번 문제입니다.

1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.

예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.

이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

이 삼각수들의 약수를 구해봅시다.

 1: 1
 
 3: 1, 3
 
 6: 1, 2, 3, 6
 
10: 1, 2, 5, 10

15: 1, 3, 5, 15

21: 1, 3, 7, 21

28: 1, 2, 4, 7, 14, 28 

위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?


```python
#프로젝트 오일러 12번 문제입니다. 소인수 분해하는 공식이 제일 중요
import math
def FindSub(power):
    sum = 1
    subnum = 0
    
    for b in range(2, int(math.sqrt(power))):
        subnum = 0
        while power % b == 0:
            power = power / b
            subnum += 1
        sum = sum * (subnum+1)
        if power == 1:
            break
        if b == 2:
            b += 1
        else:
            b += 2
        
    if power != 1:
        sum *= 2
        
    return sum

count = 0
x = 1

while count < 500:
    if FindSub((x*(x+1))/2) >= 500:
        count = FindSub((x*(x+1))/2)
        break
    x += 1
print (x*(x+1))/2
```

양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

n → n / 2  (n이 짝수일 때)

n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 

(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.


```python
#프로젝트 오일러 14번 문제입니다.
def Calculate(x, i):
    while x != 1:
        if x % 2 == 0:
            x = x / 2
            i += 1
        elif x % 2 == 1:
            x = (3 * x) + 1
            i += 1
    return i

maxnum = 0
for x in range(2, 1000001):
    i = 0
    sum = Calculate(x, i)
    if maxnum < sum:
        maxnum = sum
        order = x
    x += 1
print '가장 오래 걸리는 숫자는 : ', order, " ", '걸린 횟수는 : ', maxnum
```

우선 1을 제외한 100만까지의 모든 숫자들을 걸린 횟수를 계산하는 Calculate 함수에 대입하여 각 숫자가 얼마나 걸리는지 계산합니다.

그리고 나온 값들을 최대값과 비교하여 최대값인지 아닌지 판단해줍니다. 

Calculate 함수는 숫자를 받아서 숫자가 짝수면 2로 나누어주고, 그렇지 않으면 * 3 + 1 계산을 통하여 짝수로 바꾸어 줍니다.

이러한 과정을 통하여 진행후, 최대값을 찾아서 그 수와 같이 출력합니다.

```python
import urllib2
import string
import pickle

def mycmp(a1, a2):
    return cmp(a2, a1);

def findkey(x, lk, d):
    a = ''
    for a in range(len(d)):
        if d[lk[a]] == x:
            a = lk[a]
            break
    return a;

"""
http://www.nytimes.com
https://thevoid.com
http://www.oracle.com/index.html
http://www.android.com/intl/en_in/
https://developers.google.com/
"""
source = urllib2.urlopen("https://developers.google.com").read()

f = open('developers.google.com.html', 'w')
f.write(source)
f.close()

starting = False  #태그 여부 변수
changedsource = ''
for x in range(len(source)):
    if source[x] == '<':
        starting = True
    elif source[x] == '>':
        starting = False
    elif starting == 0 and source[x] != '<' and source[x] != '>' and source[x] != '-':
        changedsource += source[x]
        

for token in string.punctuation:
    changedsource = changedsource.replace(token, '')
        
chrless = changedsource.split()

arraydict = {}

for x in chrless:
    if arraydict.get(x) not in arraydict:
        arraydict[x] = chrless.count(x)

finalarray = arraydict.items()
finalarray.sort()

"""
for key, value in finalarray:
    print unicode(key, 'utf-8'),':', value
"""
#-------------------------------------------3번째 assignments---------------------------------------------------------
f = open('developers.google.com.frequncy.txt', 'w')

awords = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']

for a in awords:
    if arraydict.get(a):
        arraydict.pop(a)

finalarrayk = arraydict.keys()
finalarrayv = arraydict.values()
finalarrayv.sort(mycmp)

for x in finalarrayv:
    f.write(findkey(x, finalarrayk, arraydict))
    f.write(' ')
    f.write(str(x))
    f.write('\n')
    #print findkey(x, finalarrayk, arraydict), x



pickle.dump(arraydict, f)
f.close


f = open('www.nytimes.com.frequncy.txt')
i = 0
while i < 3:
    i+=1
    abc = f.readline()
    print abc
    
f.close()
```

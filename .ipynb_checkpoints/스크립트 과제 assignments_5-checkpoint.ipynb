•요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[ Count (step: %d)] %d"%(self.step,self.count)
    def incr(self):
        self.count = self.count + self.step
    def __call__(self):
        self.incr()
    def __add__(self, num):
        self.count = self.count + num
        return self.count
    def __sub__(self, num):
        self.count = self.count - num
        return self.count
    def __gt__(self,other):
        return self.count > other
    def __lt__(self,other):
        return self.count < other
    def __eq__(self,other):
        return self.count==other
    def __ne__(self,other):
        return self.count!=other

    

        
c = Counter(10)
d = Counter(10, 2)
print c
print d
print

#c증가
c.incr()
print c
c.incr()
print c
#d증가
d.incr()
print d
d.incr()
print d
d.incr()
print d
print

c() #함수를 직접 호출(call)
print c
d() #함수를 직접 호출(call)
print d
print

c = c + 100
d = d + 30

print c
print d
print

print c > 120
print c < 120
print c == 113
print d != 46
```

설명 

Counter 클래스에서 __init__, __str__, incr, __call__, __add__, __sub__, 나머지 대소 비교 및 동일 여부를 알리는 함수를 

정의하였다. 생성자 부분에서는 처음 값과, 증가하는 step값을 받는데, 이떄 step값은 디폴트 값이 1로 설정된 상태에서 받으므로, 

인자가 없는 경우는 1로 받는다. 출력 부분은 카운터와 step의 수치를 출력해주는 역할을 한다. 증가 함수에서는 초기 값을 step만큼 증가시키고, 

그 값을 초기값에 저장하는 역할을 하고, call 호출함수는 앞서 얘기한 증가함수를 다시 부르는 역할을 한다. add함수는 인스턴스에 + 연산을 

통하여 숫자를 더한 경우, 그 숫자만큼 인스턴스 객체 내의 count변수의 값을 증가시켜준다. sub는 반대로 감소시키는 역할을 하게 된다.

[4번 문제]

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 

다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 

자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

•MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.


```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

설명 : init(), str(), elimicate_duplicate()은 각각 생성자 정의, 출력 혹은 str문자열을 호출하거나 사용할때 호출, 리스트로 받은

인자들을 다시 선언한 리스트에 중복을 제거하여 삽입한다. 그리고 이 함수를 정적함수로 선언한다. 

우선 맨 처음 생성자 정의 부분을 보면 인자로 받은 리스트를 인스턴스 객체에 복사하여 elimicate_duplicate()함수에 보내주고 거기서 나온 

값을 인스턴스 객체 변수에 전달해주는 역할을 합니다.

두 번째로 str() 함수는 MySet: {인자들}의 형식으로 출력하도록 하는데, 우선 MySet:를 출력해주고 인스턴스 객체내의 리스트의 내용들을 

차례로 출력한 다음 괄호를 닫아서 마치는 함수이다.

마지막으로는 elimicate_duplicate() 함수가 있는데, 이 함수는 생성자 함수에서 받은 리스트를 다시 받아서 리스트의 내부 인자들을 

다시 꺼내서 중복 여부를 검사하고 중복되지 않은 경우만 새로 정의한 리스트에 삽입하는 과정을 거칩니다. 그리고 그 리스트를 다시 

새로운 리스트에 대입하고 생성자로 받은 인스턴스에 넘겨줍니다

[5번 문제]

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

•정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 

주석등으로 해당 코드를 설명한다.

•요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.


->u = s | t 

-> print u 
 
 MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 



•요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.


-> u = s & t 

->> print u 
 
 MySet: {2, 3} 



•요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.


-> s = MySet([1, 2, 3]) 

-> t = MySet([3, 4, 5]) 
 
-> u = s - t 
 
-> print u 
 
 MySet: {1} 



```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet : {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def __or__(self, a):
        nlist = []
        for i in self:
            nlist.append(i)
        for i in a:
            if i not in self:
                nlist.append(i)
        return nlist
    def __and__(self, a):
        nlist = []
        for i in a:
            if (i in self) and (i in a):
                nlist.append(i)
        return nlist
    def __sub__(self, a):
        nlist = []
        for i in self:
            if i not in a:
                nlist.append(i)
        return nlist
     

    

s = MySet([1, 2, 2, 3])
print s
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
print t

#중복 원소를 제거한 s, t의 합집합, 교집합, 차집합을 구하는 과정입니다.

print "합집합 : ", s|t
print "교집합 : ", s&t
print "차집합 : ", s-t
```

설명 :

우선 합집합은 연산을 실행시 or함수가 실행되도록 하였는데 이때 or 함수 내부에서 새로운 리스트를 하나 선언하고 해당 인스턴스의 

리스트의 내용들을 미리 넣어줍니다. 그리고 인자로 받은 나머지 다른 하나의 리스트이 원소들이 이미 리스트에 존재하는지 먼저 판별하고

그렇지 않으면 새롭게 추가하는 식으로 합집합을 구성하였습니다. 그리고 and 연산은 인스턴스 객체의 리스트와 인자로 받은 다른 리스트의 

원소들을 비교해서 양쪽 모두에 존재하는 리스트 원소만 뽑아내 새로운 리스트에 추가하였습니다.

마지막으로 차집합의 경우는 인스턴스 객체의 리스트 원소들 중에서 나머지 다른 리스트에 존재하는 원소들을 제거한 후 남은 원소들만 

새로운 리스트에 추가하도록 하였습니다.

1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고,

각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.

1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다.

   예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,
   
   115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.


```python
#프로젝트 오일러 17번 문제입니다.
sum = 0

array = [0, 3, 3, 5, 4, 4, 3, 5, 5, 4, 3, 6, 6, 8, 8, 7, 7, 9, 8, 8] #0부터 19까지 알파벳의 갯수 0은 값이 0
array2 = [0, 0, 6, 6, 5, 5, 5, 7, 6, 6] #10부터 90까지 알파벳 갯수 10단위 다만 10은 값이 없다.

for x in range(1, 1001):
    if x % 100 < 20:
        a = x % 100
        sum = sum + array[a]
    elif x % 100 >= 20:
        b = x % 100
        sum = sum + array[b%10] + array2[(b%100)/10]
    if (x / 100 > 0)and (x / 100) < 10:
        c = x / 100
        sum += array[c]
        sum += 7
        if x % 100 != 0:
            sum += 3
    elif x / 100 == 10:
        sum += 11
    
print sum
```

설명 : 설명에 앞서 20이상의 숫자들은 일정한 규칙을 가지고 알파벳의 단어가 오는 것을 확인 할 수 있다. 반면에 0부터 19까지는 알파벳의 

갯수를 직접 입력하여 미리 정해준 다음 코딩을 시작하였다.

우선 0부터 19까지는 따로 지정된 영어단어가 있으므로 이때는 array에서 선언한 값을 사용하여 총 합 변수에 더해준다.

그리고 20부터 100까지는 십의 자리수에 해당하는 알파벳의 갯수, 일의 자리에 해당하는 알파벳의 갯수들을 전부 더하였고, 

이 과정을 999까지 진행하였다. 

각 자리의 숫자에 해당하는 알파벳 갯수는 해당 '숫자 % 100'이 20이상이면 arrry와 array2의 배열에서 가져오도록 하였고,

20미만이면 array에서만 인덱스에 맞게 가져오도록 하였다.

그리고 마지막 1000은 one thousand이므로 총 11글자. 그러므로 총합 변수에 11을 더하여 최종적으로 값을 전부 구하였다.

프로젝트 오일러 문제 21


n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,

서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 

a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다. 

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.

또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.

따라서 220과 284는 친화쌍이 됩니다. 

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.


```python
#프로젝트 오일러 21번 문제입니다. 
def Friendly(x):
    sum1 = 0
    sum2 = 0
    returning = False
    for a in range(1, (x/2) + 1):
        if x%a == 0:
            sum1 += a
    for b in range(1, (sum1/2)+1):
        if sum1 % b == 0:
            sum2 += b
    if sum2 == x and sum2 != sum1:
        returning = True
    return returning

        

sum = 0

for x in range(2, 10001):
    if Friendly(x):
        sum += x
print sum
```

설명 : 우애수를 찾는 문제, 여기서 2부터 10000까지의 모든 수들을 고려해야 하므로, 우애수인지 판단하는 Friendly함수로 그 값을 판별하고

총합 변수에 더해주고 그 값을 출력한다.

Friendly함수는 인자로 온 숫자(a라고 한다.)에서 해당 숫자를 분해하여 나오는 숫자를 도출하고(b라고 한다.), 

b를 a와 마찬가지로 약수들을 전부 더하여 a가 나오면 그 수는 우애수로 판별하여 true를 반환하도록 한다.

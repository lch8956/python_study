```python
s = 'Hello World and Python'
l = s.split()

array  = []

for x in range(len(l)):
    array.append(s.split()[len(l) - x - 1])
    
print ''.join(array)
```

•다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 

for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

◾1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

◦>>> addall([1])

◾2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

◦>>> addallodd([1])

◦>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])

25


◾[참고]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 

사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.) 

```python
#1) 문제 코드입니다.
def addall(list):
    return sum([a for a in list]);

#2) 문제 코드입니다.
def addallodd(list):
    return sum([a for a in list if a % 2 == 1]);

print addall([3, 4, 7, 2, 4, 9, 6])
print addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

1) addall() 함수를 한줄로 작성하는 코드인데 설명은 다음과 같습니다.

우선 함수를 선언하고, 함수에 숫자값이 들어있는 리스트를 매개변수로 전달합니다. 그리고 함수에서 그 리스트를 받습니다.

그리고 함수 내에서 리스트 내포형식으로 매개변수로 받은 리스트의 정보들을 하나하나 받아서 새 리스트에 넣어줍니다.

그리고 sum() 함수를 통하여 리스트 내의 모든 값들을 더하여 그 값을 반환해줍니다.

2) addallodd() 함수를 한줄로 작성하는 코드인데 설명은 다음과 같습니다.

위의 1번 문제와 같이 리스트를 매개변수로 받고 함수 내에서 받은 리스트를 가져옵니다. 이때 리스트 내포 형식의 반복문에서 조건문을 추가합니다.

◾1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

◾2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
def mycmp(a1, a2):
    return cmp(a2, a1);

def findkey(x, lk, d):
    a = ''
    for a in range(len(d)):
        if d[lk[a]] == x:
            a = lk[a]
            break
    return a;

d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}

lk = d.keys()
lv = d.values()

lk.sort(key = str)
print '키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.'
for x in lk:
    print x, d[x]
print

lk.sort(key = str, reverse = True)
print '키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.'
for x in lk:
    print x, d[x]
print

print '값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오. '
lv.sort(cmp)
for x in lv:
    print findkey(x, lk, d), x
print

print '값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.'
lv.sort(mycmp)
for x in lv:
    print findkey(x, lk, d), x
print
```

8번 문제입니다.

소수를 크기 순으로 나열하면 2, 3, 5, 7, 11, 13, ... 과 같이 됩니다.

이 때 10,001번째의 소수를 구하세요.

```python
#프로젝트 오일러 7번 
import math

def isPrime(x): 
    returning = True 
    for i in range(2,int(math.sqrt(x))+1):
        if x%i==0: 
            returning = False
            return returning
    return returning

array = []
x = 1
while len(array) <= 10001:
    if isPrime(x):
        array += [x]
    x += 1
print array[-1]
```

우선 10001번째 소수가 언제 오는지 정확하게 알 수 없기 떄문에 while()문을 사용하였고 (소수를 저장하는 리스트의 크기가 10001를 넘기면

반복문 탈출), isPrime() 함수를 이용하여 소수를 판별하도록 하였습니다. 1부터 숫자를 1씩 증가시키면서 그 때마다 소수인지 아닌지 판별하고,

소수이면 앞서 미리 선언한 빈 리스트에 소수값을 하나씩 추가해줍니다. 그리고 그 값이 10001개가 되면 반복문을 빠져나와서

예를 들면 32 + 42 = 9 + 16 = 25 = 52이므로 3, 4, 5는 피타고라스 수입니다.

a + b + c = 1000  인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
#프로젝트 오일러 9번 문제입니다.
output = 0
stop = False
for a in range(1, 1001):
    for b in range(1, 1001):
        for c in range(1, 1001):
            if a + b + c != 1000:
                continue
            elif a + b + c == 1000:
                if a*a + b*b == c*c:
                    print a, b, c
                    stop = True
                    output = a*b*c
                    break
                else:
                    continue
        if stop == True:
                break
    if stop == True:
                break
print output
```

우선 3중 반복문을 선언하여서 각 반복문이 1000번씩 수행되도록 하였습니다. 그리고 반복문 내에서 수해오디는 각 인자들의 

합이 1000이 되는 경우, 그리고 거기서 피타고라스의 법칙  a2 + b2 = c2 을 만족하는 a, b, c가 있는 경우 그 값들을 전부 곱한 값을 

출력하도록 합니다.

◾요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.

◦예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']

◦모든 punctuation을 확인하는 방법은 아래와 같다

◾요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.

◾요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.

◦http://www.naver.com

◦http://www.daum.net

◦http://www.nytimes.com

```python
import urllib2
import string

source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

starting = False  #태그 여부 변수
changedsource = ''
for x in range(len(source)):
    if source[x] == '<':
        starting = True
    elif source[x] == '>':
        starting = False
    elif starting == 0 and source[x] != '<' and source[x] != '>' and source[x] != '-':
        changedsource += source[x]
        

for token in string.punctuation:
    changedsource = changedsource.replace(token, '')
        
chrless = changedsource.split()

arraydict = {}

for x in chrless:
    if arraydict.get(x) not in arraydict:
        arraydict[x] = chrless.count(x)

finalarray = arraydict.items()

for key, value in finalarray:
    print unicode(key, 'utf-8'),':', value
```
